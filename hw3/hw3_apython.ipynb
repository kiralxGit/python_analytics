## Задача 1
Скачать данные по ссылке https://drive.google.com/file/d/1Ep9ncT58xJ059I3hDeX2YuJTq4advJTT/view

Считать данные с помощью pandas

Вывести на экран первые 5 строк

```python
houses = pd.read_csv('kc-house-data.csv')
houses.head()
```

### 1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади


```python
houses['price_per_sq_lot'] = houses.apply(lambda x:
                                          x['price'] / x['sqft_lot'],
                                          axis = 1).round(2)
houses.head()
```

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
houses['delta_renovated'] = houses.apply(lambda x:
                                         0 if x['yr_renovated'] == 0
                                         else x['yr_renovated'] - x['yr_built'],
                                         axis = 1)

houses.head()
# houses[houses['delta_renovated'] > 0].head(10)
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
houses['date'] = houses['date'].astype("datetime64[ns]")
houses['year_of_sale'] = houses['date'].dt.year
houses['month_of_sale'] = houses['date'].dt.month 
houses.head()
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
houses.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
houses.head()
```

## Задача 3

Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален

Отсортируйте от меньшей стоимости к большей

```python
houses.pivot_table(index='bedrooms', values='price', aggfunc='mean').sort_values(by=['price'])
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
houses.pivot_table(index='condition', aggfunc={'price': ['min', 'mean', 'max']})
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
pd.crosstab(index=houses['view'], columns=houses['waterfront'])
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=houses['bedrooms'], columns=houses['floors'])
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
pd.crosstab(index=houses['grade'],
            columns=houses['condition'],
            values=houses['price'],
            aggfunc='median').fillna('-')
```
